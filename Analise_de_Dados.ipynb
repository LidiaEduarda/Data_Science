```python
import pandas as pd
```

```python
# Importanto meu banco de dados das taxas de taxi que está no formato csv, usando o metodo ( pd.read_cdv() - Importa os dados)
taxis = pd.read_csv('taxis.csv')
taxis.head() # Metodo (.head() - Para mostar apenas as 5 primeira linhas dos dados)
```

```python
# Visualizando na coluna de vizinha a quantidade de vezes dos locais onde o passageiro foi deixado, usando a função ( .value_counts() - Mostra a quantidade de vezes que o elemento se repete).
taxis['dropoff_borough'].value_counts()
```

```python
# Visualizando se as cincos vizinhaças em que os valores das casas custuman ser mais caras, oferecem melhores gorgetas, usando a função ( .groupby([])[] - Agrupa os elementos que voce deseja).
# Após o metodo descrito acima,  estou passando algumas agregaçoes para ele realizar, usndo a função ( .agg([,]) - Faz agregaçoes de alguns metodo como media e mediana).
taxis.groupby(['dropoff_borough'])['tip'].agg(['mean', 'median'])
```
