```python
import pandas as pd

!pip install mlxtend

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
from mlxtend.plotting import plot_confusion_matrix
from sklearn.metrics import confusion_matrix, classification_report
```

```python
df = pd.read_csv('breast-cancer.csv')
```

```python
df.head()
```

```python
x = df.drop(['id', 'diagnosis'], axis=1) # passando uma lista de colunas que desejo excluir com o método drop, armazenando na variavel x.
x
```

```python
# Passando a coluna ('diagnosis') para a variavel y
y = df['diagnosis']
y
```

```python
modelo = LogisticRegression(max_iter=10000)
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y) 
```

```python
modelo.fit(x_treino, y_treino)
```

```python
y_previsto = modelo.predict(x_teste)
```

```python
cm = confusion_matrix(y_previsto, y_teste)
```

```python
plot_confusion_matrix(cm) # c
```

```python
print(classification_report(y_previsto, y_teste))
```
