```python
import pandas as pd
import matplotlib.pyplot as plt # Biblioteca  utilizada para criar visualizações de dados, como gráficos de linhas, gráficos de barras, histogramas, dispersão etc. Ao importar a biblioteca e renomeá-la como plt, torna-se mais conveniente usar suas funções, pois podemos chamar as funções diretamente através de plt
```

```python
# Visualizando meu DataSet
df = pd.read_csv('Best Movies Netflix.csv')
df.head() 
```

```python
# Criando variaveis que armazem os nomes e as quantidades quantidas na coluna ('MAIN_GENRE')
# A função (.value_counts() conta quantas vezes cada gênero aparece na coluna MAIN_GENRE ).
eixo_x = df['MAIN_GENRE'].value_counts().keys() # .keys() extrai os nomes dos gêneros (rótulos).
eixo_y = df['MAIN_GENRE'].value_counts().values # .values extrai as contagens de cada gênero (números).
```

```python
# Plotando meu grafico
plt.bar(eixo_x, eixo_y) # ( plt.bar(): Cria um gráfico de barras, onde cada categoria no eixo x possui uma barra com altura correspondente ao valor no eixo y. É útil para representar dados discretos, como contagens ou frequências de categorias ).


# Estilizando meu grafico
plt.title('Filmes por Genero') # ( plt.title('Filmes por Genero'): Esta linha adiciona um título ao gráfico, que neste caso é "Filmes por Gênero") .
plt.xlabel('Generos de Filme') # ( plt.xlabel('Generos de Filme'): Esta linha adiciona um rótulo ao eixo x do gráfico, que neste caso é "Gêneros de Filme" ).
plt.ylabel('Quantidade de Filmes') # ( plt.ylabel('Quantidade de Filmes'): Esta linha adiciona um rótulo ao eixo y do gráfico, que neste caso é "Quantidade de Filmes" ).
plt.xticks(rotation=45) # ( plt.xticks(rotation=45): Esta linha rotaciona os rótulos das categorias no eixo x em 45 graus ). 


plt.show() # ( plt.show(): Esta linha exibe o gráfico completo na tela. Após todas as configurações e adições de dados, essa função é chamada para mostrar o gráfico final ).
```

```python
# Analise: Os filmes do tipo drama são os que mais aparecem na lista de melhores filmes, pois eles tem notas bem altas. Provavelmente a netflix vai querer investir neles.
```

```python
# Criando variaveis que armazem os anos e a quantidade de vezes que ele aparaceu na coluna ('RELEASE_YEAR')
# ( .sort_index() é um método utilizado em pandas para ordenar um DataFrame ou uma Series pelos índices. Ele reorganiza os dados com base nos índices das linhas ou das colunas, dependendo de onde é aplicado ).
eixo_x = df['RELEASE_YEAR'].value_counts().sort_index().keys() 
eixo_y = df['RELEASE_YEAR'].value_counts().sort_index().values 
```

```python
# Plotando meu grafico
plt.plot(eixo_x, eixo_y) # ( plt.plot(): Essa função cria um gráfico de linhas, onde cada par de valores em eixo_x e eixo_y é conectado por uma linha reta ).


# Estilizando meu grafico
plt.title('Melhores Filmes por Ano de Lançamento') 
plt.xlabel('Ano de Lançamento')
plt.ylabel('Quantidade de Filmes')


plt.show()
```

```python
# Analise: Tem uma evolução na quantidade de filmes dentre esses filmes de melhores avaliados nos ultimos anos. Isso pode se dever ao fato de terem mais filmes nesses anos etc.
```
