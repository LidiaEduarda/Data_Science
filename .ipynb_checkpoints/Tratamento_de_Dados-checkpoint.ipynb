```python
df = pd.read_excel('Sao paulo.xlsx')
df
```

```python
# A função ( .loc[] - É usada para acessar um grupo de linhas ou colunas ).
# Criando uma Variavel e armazenando a linha (10031 do meus dados)
limpeza_linha = df.loc[10031]
limpeza_linha
```

```python
# criando uma função para auxiliar a modificar a alinha [10031]
# A função (.replace() - E usada para substituir valores em uma DataFrame ou Series por outros valores. )
# Estou repetindo o que o parametro da função  ira fazer pois acho que nao pode passar todas as mudanças em uma unica função (.replace())
def limpa_preco(linha):
    linha['price'] = linha['price'].replace('R$ ', '') # Linha = Linha, no lugar do (R$ ) estou substituindo por nada
    linha['price'] = linha['price'].replace('\n', '') # Linha = Linha, no lugar do (\n) estou substituindo por nada
    linha['price'] = linha['price'].replace('/Mês', '') # Linha = Linha, no lugar do (/Mês) estou substituindo por nada
    linha['price'] = linha['price'].replace('                      ', '') # Linha = Linha, no lugar do (                      ) estou substituindo por nada
    linha['price'] = linha['price'].replace('/Ano', '') # Linha = Linha, no lugar do (/Ano) estou substituindo por nada
    linha['price'] = linha['price'].replace('/Dia', '') # Linha = Linha, no lugar do (/Dia) estou substituindo por nada
    return linha
```

```python
# Verificando a função (limpa_preco), e para ver o que ele mudou chamo a variavel limpiza_linha pois  nela que está armazenda todas informações d linha 
limpa_preco(limpeza_linha)
```

```python
# A função apply() do pandas é usada para aplicar uma função ao longo de um eixo de um DataFrame. Neste caso, apply() está sendo utilizado para aplicar a função limpa_preco a cada linha.
# Um lambda é uma função anônima em Python, que neste caso recebe um argumento x.
# Aqui, x representa cada linha do DataFrame df.
# O argumento axis=1 indica que a função deve ser aplicada ao longo das linhas. Se fosse axis=0, a função seria aplicada ao longo das colunas.
df.apply(lambda x: limpa_preco(x), axis=1)
```

```python
def ajusta_preco_alugueis(linha):
    linha['price'] = linha['price'].replace('.', '') # Linha = Linha, no lugar do (.) estou substituindo por nada
    preco = int(linha['price'])
    if (preco < 10000):
        preco = preco * 200
    linha['price'] = preco
    return linha
```

```python
# ajusta_preco_alugueis(limpeza_linha) -> Nao rode este codogo pois ele vai alter a linha novamente que vc ja converteu para inter 
```

```python
# Aplicando a funcao (ajusta_preco_alugueis) para o resto do meu DateFrame 
df.apply(lambda x: ajusta_preco_alugueis(x), axis=1)
```
